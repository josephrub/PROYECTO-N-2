```python
'''
- Generar 5 nuevas variables aplicando los siguientes criterios
con las variables numéricas:
1. Log10(var1/var2)
2. Sqrt(var1)*exp(var2)/200
3. Si var 1 > var 2 → 5, caso contrario 3
4. 1/logn(var1/var2)*100
5. var2**2/var1
'''
```

```python
var1="Desplazamiento"
var2="Aceleracion"


var1=np.array(df["Desplazamiento"])
var2=np.array(df["Aceleracion"])

#1. Log10(var1/var2)
col1=np.log10(var1/var2)

#2. Sqrt(var1)*exp(var2)/200
col2=(np.sqrt(var1)*np.exp(var2))/200

#3. Si var 1 > var 2 → 5, caso contrario 3
col3=list()
for i in range(0,len(df["Desplazamiento"])):
  if var1[i] > var2[i]:
    col3.append(5)
  else:
    col3.append(3)
  
#4. 1/logn(var1/var2)*100
col4=(1/(np.log(var1/var2)))*100

#5. var2**2/var1
col5= (var2**2)/var1


df["Columna 1"]= list(col1)
df["Columna 2"]= list(col2)
df["Columna 3"]= list(col3)
df["Columna 4"]= list(col4)
df["Columna 5"]= list(col5)


df.head(20)
```

```python
'''
Ejercicio No 3

Crear una función que permita generar 3 variables
cualitativas ordinales a partir de las variables numéricas con
los siguientes criterios:
★ Si es mayor a la “media + 1*desviación estándar”
→ Alto
★ Si está entre “media - 1*desviación estándar” y
“media + 1*desviación estándar” → Medio
★ Si es menor a la “media - 1*desviación estándar”
→ Bajo
'''
```

```python
def categoria(d):
  prom=d.mean()
  desv=d.std()
  c=list()

  for i in range(0,len(d)):
    if d[i] > prom + desv:
      c.append("Alto")
    elif d[i] >= prom-desv and d[i] <= prom+desv:
      c.append("Medio")
    else:
      c.append("Bajo")
  return c
```

```python
df["Categoria"]=categoria(df["Peso_En_Libras"])
df.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def grafico(cuanti,cuali):
  _, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,9))
  
  sns.boxplot(x=cuali, y=cuanti, data=df, ax=axes[0])
  sns.barplot(x=cuali, y=cuanti, estimator=np.mean, data=df, ax=axes[1])
  sns.displot(data=df, x=cuanti, hue=cuali, kde=True)
```

```python
grafico("Millas_Por_Galon","Pais_Origen")
grafico("Peso_En_Libras","Pais_Origen")
```
